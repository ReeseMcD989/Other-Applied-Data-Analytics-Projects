### Explanation of PCA by ChatGPT:

PCA, short for Principal Component Analysis, is a widely used technique in data analysis and machine learning for dimensionality reduction and feature extraction. It aims to transform a high-dimensional dataset into a lower-dimensional representation while retaining as much of the original information as possible. PCA accomplishes this by identifying the principal components of the data, which are linear combinations of the original features.

Here's a high-level overview of how PCA works:

1. Standardization: PCA typically begins with standardizing the features of the dataset to have zero mean and unit variance. This step ensures that all features contribute equally to the analysis and prevents any single feature from dominating the results.

2. Covariance Matrix: PCA constructs the covariance matrix from the standardized data. The covariance matrix describes the relationships between different features by measuring their pairwise covariances.

3. Eigendecomposition: PCA performs an eigendecomposition of the covariance matrix. This decomposition yields a set of eigenvalues and eigenvectors. The eigenvalues represent the amount of variance explained by each eigenvector (principal component), while the eigenvectors define the directions in the original feature space along which the data vary the most.

4. Selection of Principal Components: PCA sorts the eigenvalues in descending order and selects the top k eigenvectors corresponding to the largest eigenvalues. These eigenvectors are the principal components. The number of principal components chosen, k, determines the dimensionality of the reduced dataset.

5. Projection: The selected principal components are used to create a new transformed dataset by projecting the original data onto the subspace defined by these components. Each data point in the transformed dataset represents its original data expressed in terms of the principal components.

The benefits of PCA include:

- Dimensionality Reduction: PCA reduces the number of features while preserving the most important information, making it useful for datasets with high dimensionality.
- Feature Extraction: The principal components obtained from PCA can be seen as new features that capture the most significant patterns in the data.
- Visualization: By reducing the dimensionality, PCA enables visual representation of data in lower-dimensional space, allowing easier interpretation and analysis.
- Noise Reduction: PCA can filter out noise or irrelevant variations in the data by emphasizing the principal components that capture the most important patterns.

PCA has various applications, including exploratory data analysis, image and signal processing, feature engineering, and pattern recognition. It is a fundamental tool in many fields where dealing with high-dimensional data is challenging or requires efficient representation.

### Explanation of principle components:

Principal components are the linear combinations of the original features in a dataset that capture the maximum amount of variance in the data. Each principal component is a new feature that represents a direction in the original feature space along which the data vary the most.

To better understand principal components, let's consider an example with a dataset of customers and their purchasing behavior:

Suppose we have the following original features:
- Age: Age of the customer
- Income: Annual income of the customer
- Spending: Amount spent by the customer on shopping
- Visits: Number of visits to the store

After applying PCA to this dataset, we obtain principal components, which are combinations of these original features. Let's say we have two principal components, PC1 and PC2.

Example of Principal Component 1 (PC1):
PC1 could be a linear combination of Age, Income, and Spending. It represents a direction in the original feature space that captures the most significant variation in the data. PC1 might indicate the overall spending power of customers, reflecting a combination of age, income, and spending habits.

Example of Principal Component 2 (PC2):
PC2 could be a linear combination of Age and Visits. It represents another direction in the feature space that captures additional variation not explained by PC1. PC2 might reveal a pattern related to customer engagement or loyalty, as it combines information about age and visit frequency.

Each subsequent principal component explains less variance than the previous one, capturing progressively smaller patterns in the data. The total number of principal components is equal to the number of original features in the dataset.

It's important to note that the interpretation of principal components depends on the specific dataset and domain knowledge. PCA provides a mathematical representation of the most significant patterns in the data but does not necessarily assign specific meanings to each principal component. Interpreting the principal components often requires analyzing the relationships between the original features and the context of the problem being solved.

### Variance in the context of PCA:

In the context of PCA (Principal Component Analysis), variance refers to the spread or variability of the data along a particular axis or direction in the feature space. It measures how much the values in a dataset deviate from their mean or average value.

Variance is a statistical measure that quantifies the dispersion or spread of a set of observations. In PCA, variance is used to capture the most important patterns and variability in the data. The principal components are selected based on the amount of variance they explain.

In PCA (Principal Component Analysis), the explained variance refers to the amount of variance in the original data that is accounted for by each principal component. It quantifies the proportion of the total variance explained by each principal component.

When performing PCA, the principal components are ranked in descending order based on the amount of variance they explain. The first principal component explains the largest amount of variance, the second principal component explains the second-largest amount, and so on. The explained variance can be expressed as a ratio or percentage.

The explained variance provides insights into the importance or significance of each principal component. Principal components with higher explained variance contribute more to the overall structure and patterns in the data. By examining the explained variance of each component, one can determine the optimal number of principal components to retain for dimensionality reduction or feature extraction.

Additionally, the cumulative explained variance is also useful to evaluate how much of the total variance is explained by a given number of principal components. By calculating the cumulative sum of the explained variances, one can determine the point at which the cumulative explained variance reaches a satisfactory level, indicating the number of principal components to retain.

In practice, it is common to plot the explained variance or cumulative explained variance to visualize the contribution of each principal component and determine the appropriate dimensionality reduction for the specific problem at hand.

### Reading in the data with a query of GBQ data base.

```python
# SELECT customer_id, il.beverage, SUM(gross_sales) AS sum_gross_sales: Selects the customer_id, il.beverage, and the sum 
    # of gross_sales from the specified tables.
    
# FROM umt-msba.dram_shop.dram_items_* AS di: Specifies the source table dram_items_* with a wildcard to match multiple 
    # tables.
    
# JOIN umt-msba.dram_shop.item_lookup AS il ON di.item = il.item: Joins the dram_items_* table with the item_lookup table 
    # on the item column.
    
# WHERE customer_id IS NOT NULL AND il.beverage != "custom amount" AND il.beverage IN (SELECT beverage FROM 
    # reese-msba.dram_shop.dram_items_*AS di JOINumt-msba.dram_shop.item_lookup AS il ON di.item = il.item WHERE 
    # beverage IS NOT NULL GROUP BY beverage ORDER BY SUM(gross_sales) DESC LIMIT 1000): Specifies filtering 
    # conditions for the query. It includes conditions for customer_id not being null, il.beverage not equal 
    # to "custom amount", and il.beverage being in a subquery result.
    
# GROUP BY customer_id, il.beverage: Groups the results by customer_id and il.beverage.

# ORDER BY sum_gross_sales DESC: Orders the results by the sum_gross_sales column in descending order.

query = '''
SELECT
  customer_id,
  il.beverage,
  SUM(gross_sales) AS sum_gross_sales
FROM
  `umt-msba.dram_shop.dram_items_*` AS di
JOIN
  `umt-msba.dram_shop.item_lookup` AS il
ON
  di.item = il.item
WHERE
  customer_id IS NOT NULL
  AND il.beverage != "custom amount"
  AND il.beverage IN (
  SELECT
    beverage
  FROM
    `reese-msba.dram_shop.dram_items_*` AS di
  JOIN
    `umt-msba.dram_shop.item_lookup` AS il
  ON
    di.item = il.item
  WHERE
    beverage IS NOT NULL
  GROUP BY
    beverage
  ORDER BY
    SUM(gross_sales) DESC
  LIMIT
    1000 )
GROUP BY
  customer_id,
  il.beverage
ORDER BY
  sum_gross_sales DESC
'''

# This line executes a query on the Google BigQuery client. It calls the query method of the client object, passing in the 
    # query parameter, and assigns the result to the beverages variable.
    
beverages = client.query(query)

# This line executes a query on Google BigQuery using the pandas_gbq library. It calls the read_gbq function, passing in 
    # the query and project_id parameters, and assigns the result to the beverages variable. The project_id parameter is 
    # set to the gbq_proj_id value.

beverages = pandas_gbq.read_gbq(query,project_id = gbq_proj_id)
```

```python
# Transpose the dataframe. This line uses the pivot() method to reshape the data, setting the 'customer_id' column as the 
    # index, 'beverage' column as the columns, and 'sum_gross_sales' column as the values. The resulting pivot table is 
    # assigned to the variable bev_pivot.

bev_pivot = beverages.pivot(index='customer_id', columns = 'beverage', values='sum_gross_sales')
```

```python
# Fill na with 0.

bev_pivot = bev_pivot.fillna(0)
```

### PCA on Dram product sales

```python
# Simplify the PCA function access. This line creates an instance of the PCA (Principal Component Analysis) class. PCA is 
    # a dimensionality reduction technique used to transform high-dimensional data into a lower-dimensional representation.
    
pca = PCA()
```

```python
# Fit the PCA function to bev_pivot. This line fits the PCA model to the bev_pivot pivot table. The fit() method is used 
    # to estimate the principal components based on the data in bev_pivot. After this line, the PCA model will be ready 
    # to perform dimensionality reduction or other analysis tasks based on the fitted model.

pca.fit(bev_pivot)
```

```python
# Showing explained variance for the first 20 components.

# This line creates a DataFrame called explained_variance from the explained variance ratios obtained from the pca object. 
    # The explained_variance_ratio_ attribute of the PCA object returns the proportion of variance explained by each 
    # principal component. Each value represents the amount of variance explained by a particular principal component.

explained_variance = pd.DataFrame(pca.explained_variance_ratio_)

# This line creates a bar plot using the plot.bar() function on the first 20 rows of the explained_variance DataFrame. It 
    # assigns the resulting plot to the variable ax. The legend=False parameter removes the legend from the plot, and the 
    # figsize=(8, 4) parameter sets the size of the figure to be 8 units wide and 4 units tall.

ax = explained_variance.head(20).plot.bar(legend=False, figsize=(8, 4))

# Adjusting the plot layout, labeling the x-axis, and showing the plot.

ax.set_xlabel('Component')
plt.tight_layout()
plt.show()
```

```python
# This line creates a DataFrame called cume_explained_variance by taking the cumulative sum of the explained variance 
    # ratios obtained from the pca object. The explained_variance_ratio_ attribute returns the proportion of variance 
    # explained by each principal component. By applying cumsum(), each value in the DataFrame represents the cumulative 
    # explained variance up to that specific principal component.
    
cume_explained_variance = pd.DataFrame(pca.explained_variance_ratio_).cumsum()

# This line creates a line plot using the plot.line() function on the first 150 rows of the cume_explained_variance 
    # DataFrame. It assigns the resulting plot to the variable ax. The legend=False parameter removes the legend from 
    # the plot, and the figsize=(8, 4) parameter sets the size of the figure to be 8 units wide and 4 units tall.

ax = cume_explained_variance.head(150).plot.line(legend=False, figsize=(8, 4))

# # Adjusting the plot layout, labeling the x-axis, and showing the plot.

ax.set_xlabel('Component')
plt.tight_layout()
plt.show()
```

### Loadings

In the context of PCA (Principal Component Analysis), the PCA loadings represent the coefficients or weights assigned to each original feature in the construction of the principal components. The loadings indicate the contribution of each feature to the variation captured by each principal component.

Looking at the provided example table, the rows represent different beverages, and the columns represent the principal components. Each cell in the table corresponds to the loading of a particular beverage on a specific principal component.

The loadings quantify the relationship between the original features and the principal components. They indicate the direction and magnitude of the impact of each feature on the corresponding principal component. Positive or negative values indicate the direction of the relationship, while the magnitude represents the strength of the contribution.

For instance, let's consider the "growler - 64 oz miir stainless" beverage and the first principal component (PC1). The loading value of -0.000100 suggests that this beverage has a negative influence on PC1. Similarly, for the "montana bottle opener" beverage and PC1, the loading value of -0.000060 indicates a negative impact.

The loadings help to interpret and understand the relationship between the original features and the principal components. Features with higher absolute loading values have a stronger influence on the corresponding principal component. Features with loading values close to zero have little impact or are less relevant in capturing the variation represented by that principal component.

By examining the loadings across the principal components, one can identify which original features contribute the most to the overall variation in the dataset. Loadings with large absolute values indicate features that have a significant influence on multiple principal components, capturing important patterns in the data.

Analyzing the PCA loadings can provide insights into the key features driving the variability in the dataset. It helps in feature selection, identifying important dimensions for data representation, or understanding the underlying factors contributing to the observed patterns.

```python
# Showing how each product loads on each component.

# In this line of code, a DataFrame named "loadings" is created. It is constructed using the components of the PCA model 
    # (pca.components_) as the data and the column names of the "bev_pivot" DataFrame (bev_pivot.columns[:]) as the column 
    # labels.

loadings = pd.DataFrame(pca.components_, columns=bev_pivot.columns[:])

# In this line of code, the "loadings" DataFrame is transposed using the transpose() function. This operation swaps the 
    # rows and columns of the DataFrame. Then, the sort_values() function is applied to sort the rows of the transposed 
    # DataFrame based on the values in the first column (column index 0). This sorting is done in ascending order.

loadings.transpose().sort_values(0)
```

```python
# Plotting the 20 largest loadings of the first component.

# In this line of code, a Series named "pc_loading" is created. It is obtained by selecting the row with index 0 from 
    # the "loadings" DataFrame using the loc[] function. Then, the sort_values() function is applied to sort the values 
    # in ascending order based on the first column (column index 0). The tail() function is used to select the last 20 
    # values from the sorted Series.

pc_loading = loadings.loc[0, :].sort_values(0).tail(n=20)

# This line of code generates a bar plot for the "pc_loading" Series. Each value in the Series represents the loading of 
    # a specific beverage on the first principal component (PC1). The plot will display vertical bars representing the 
    # magnitude of the loadings.

pc_loading.plot.bar()

# These lines of code set the label for the y-axis of the plot as "PC1 Loading" using the ylabel() function. The 
    # tight_layout() function is called to automatically adjust the padding and spacing of the plot. Finally, the show() 
    # function displays the plot on the screen.

plt.ylabel("PC1 Loading")
plt.tight_layout()
plt.show()
```

```python
# Defines function that plots a chosen component

# This line defines a function named show_component that takes two arguments: component and cutoff. The component argument 
    # represents the principal component number, and the cutoff argument specifies the threshold for filtering the loadings.

def show_component(component,cutoff=0.05) : 
    
    # This line selects the column corresponding to the specified component from the transposed loadings DataFrame. It uses 
        # the iloc function to locate the column by its index, which is component-1 since indexing starts from 0.
    
    pc_loading = loadings.transpose().iloc[:,component-1]
    
    # This line filters the loadings based on their absolute values. It keeps only the loadings that exceed the specified 
        # cutoff value. The abs() function is used to compute the absolute values of the loadings. The resulting filtered 
        # loadings are then sorted in ascending order using the sort_values() function.
    
    pc_loading = pc_loading.loc[pc_loading.abs() > cutoff].sort_values()
    
    # This line calculates the maximum absolute value among the filtered loadings. It is used later to set the y-axis 
        # limits of the plot.
    
    max_pc = max(pc_loading.abs())
    
    # This line creates a list named colors that assigns colors to the loadings based on their signs. If a loading is 
        # positive (greater than 0), it is assigned color 'C0', otherwise, it is assigned color 'C1'.
    
    colors = ['C0' if l > 0 else 'C1' for l in pc_loading]
    
    # This line generates a bar plot for the filtered loadings. The color parameter is set to the colors list, so the 
        # bars will have different colors based on their signs.
    
    pc_loading.plot.bar(color = colors)
    
    # These lines modify the plot's axes. plt.gca() returns the current Axes instance, which is assigned to the ax 
        # variable. plt.axhline() is used to draw a horizontal line at y=0 to indicate the zero loading reference. 
        # ax.set_ylim() sets the limits of the y-axis to match the range of loadings (-max_pc to max_pc).
    
    ax = plt.gca()
    plt.axhline(y=0,c="#888888")
    ax.set_ylim(-max_pc, max_pc)
    
    # These lines adjust the layout of the plot, rotate the x-axis tick labels by 45 degrees, align them to the right, 
        # and display the plot on the screen using plt.show().
    
    plt.tight_layout()
    plt.xticks(rotation=45, ha='right')
    plt.show()
    
# The show_component() function provides a convenient way to visualize the loadings of a specific principal component, 
    # highlighting the loadings that exceed the specified cutoff. The resulting plot shows the loadings as vertical bars, 
    # with different colors indicating positive and negative loadings.
```
